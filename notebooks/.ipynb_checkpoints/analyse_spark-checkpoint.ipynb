```python
df = spark.read.option("multiline", "true").json("/data/emploi_dakar.json")
```

```python
# Apres la lecture de nos données, on va les stockés dans cache pour ne pas avoir a les rappeler encore.
df.cache()
```

```python
df.select("poste").count()
```

```python
df.printSchema()
```

```python
from pyspark.sql.functions import col 

df.groupBy("poste").count().orderBy(col("count").desc()).show(5)
```

```python
df.groupBy("region").count().orderBy(col("count").desc()).show(5)
```

```python
df.groupBy(df.columns).count().filter("count > 1").show()
```

```python
from pyspark.sql import functions as F 

df_clean = df.withColumn(
      "region_clean",
      F.trim(
          F.regexp_replace(F.col("region"), r"[\n\t\t\t]" , ""))
)

df_clean = df_clean.drop('region')

df_clean.show(5)
```

```python
df_clean.describe()
```

```python
df_clean.groupBy('date_de_publication').count().show()
```

```python
df_clean.filter('date_de_publication is null').show(5)
```

# Spark avec SQL_queries

```python
# A partir de notre dataFrame, on cree une view pour exxecuter du code Spark_sql
df.createOrReplaceGlobalTempView("tab_df")
```

```python
# verification des tables créées 
spark.catalog.listTables("global_temp")
```

```python
spark.sql(""" SELECT * FROM global_temp.tab_df""").show(5)
```

```python
spark.sql(""" SELECT * FROM global_temp.tab_df WHERE date_de_publication is Null """).show(5)
```

```python
spark.sql("""
    SELECT COUNT(*) AS nb_nulls
    FROM global_temp.tab_df
    WHERE date_de_publication IS NULL
""").show()
```

```python
cols = ", ".join(df.columns)
cols
```

```python
cols = ", ".join(df.columns)

spark.sql(f"""
     SELECT {cols}, count(*) AS nombre_de_duplication FROM global_temp.tab_df
     GROUP BY {cols}
     HAVING nombre_de_duplication > 1 
    """).show(5)
```

```python
spark.sql("""
      SELECT * FROM global_temp.tab_df 
      WHERE date_de_publication is Null
      """).show(5)
```
